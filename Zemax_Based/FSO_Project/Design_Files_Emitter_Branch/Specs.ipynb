# **Gaussian Parameters**

1. Source beam size: $5.2 \mu m$
2. Wavelength: $1.55 \mu m$
3. Position (Non-seq): $0.055 mm$

In Zemax Non-Sequential Mode, A Source Gaussian has two additional parameters: Beam Size and Positions.

The **Position** parameter is the distance. along the local Z axis, between the location of the source waist and the apparent divergence location of the rays. This is used to **set the divergence angle**



$$ \tan\theta = \frac{\omega(z)}{z} \quad \Rightarrow \quad z(\text{Position}) = \frac{\omega(z)}{\tan\theta} $$
