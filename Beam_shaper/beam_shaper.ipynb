## **Geometric Optical Laws of Intensity**

**Conservation of Energy**

The geometrical optics law of intensity requires the intensity times a cross-sectional area of bundle of rays be constant:

$$ \int_0^r\int_0^{2\pi} I_{in}(r)rdrd\theta = \int_0^R\int_0^{2\pi} I_{out}(R)RdRd\theta \tag{1.1} $$

where $r$ and $R$ are the ray heights of a ray at the entrance and exit aperture. 

And the input beam is a Gaussian beam which has a normalized **Intensity Profile**:

$$ I_{in}(r) = \frac{2}{\pi \omega_0^2}\exp(\frac{-2r^2}{\omega_0^2}) \tag{1.2} $$

The output beam is a normalized top hat (TH) beam:

$$ I_{out}(R) = \frac{\text{circ}(R/R_{max})}{\pi R_{max}^2} \tag{1.3} $$

where

$$
\text{circ}(\frac{R}{R_{max}}) = \begin{cases} 1 & \text{if } |\frac{R}{R_{max}}| < 1 \\ 0 & \text{if } |\frac{R}{R_{max}}| \ge 1  \end{cases} \tag{1.4}
$$

The total energy that passes of the Gaussiam beam which passes the first lens can be derived as:

$$\int^{r_{max}}_0 \int_0^{2\pi} I_{in}(r)rdrd\theta = 1 - \exp(\frac{-2r_{max}^2}{\omega_0^2}) \tag{1.5} $$

where $r_{max}$ and $R_{max}$ are the **Semi-diameter** of the entrance and exit aperture. and since the Gaussian beam profile has been normalised, this result also represent the percentage of the total power.

Therefore, the Equation 1.3 should become:

$$  I_{out}(R) = \left[1 - \exp(\frac{-2r_{max}^2}{\omega_0^2})\right]\frac{\text{circ}(R/R_{max})}{\pi R_{max}^2} \tag{1.6} $$

Substitute it in the Equation 1.1, we will have:

$$1 - \exp(\frac{-2r^2}{\omega_0^2}) = \left[1 - \exp(\frac{-2r_{max}^2}{\omega_0^2})\right] \int_0^R\int_0^{2\pi} I_{out}(R)RdRd\theta \tag{1.7} $$

Which results in the **ray mapping function** for an input Gaussian beam output top-hat beam profile:

$$R = \epsilon R_{\text{max}} \sqrt{\frac{1 - \exp[-2(r / w_0)^2]}{1 - \exp[-2(r_{\text{max}} / w_0)^2]} \tag{1.8}}$$

And inversely:

$$r = \varepsilon w_0 \times \sqrt{-\left(\frac{1}{2}\right) \ln \left\{ 1 - \left(\frac{R}{R_{\text{max}}}\right)^2 \left[ 1 - \exp \left( -2 \left(\frac{r_{\text{max}}}{w_0}\right)^2 \right) \right] \right\} } \tag{1.9}$$

where $\epsilon$ is defined by:

$$\epsilon = \begin{cases} +1 & \text{Galilean configuration} \\ -1 & \text{Keplerian configuration} \end{cases}  \tag{2}$$

$R$ can be wirtten as:

$$R = B \sqrt{1 - \exp\left[ -2 \left( \frac{r}{w_0} \right)^2 \right]} \tag{3}$$

where:

$$B = \frac{\epsilon R_{\text{max}}}{\sqrt{1 - \exp[-2(r_{\text{max}} / w_0)^2]}} \tag{4}$$



## **Aberration Theory Results**



$$\tan \alpha = \frac{r - R}{d} = \frac{r}{s'} \Rightarrow s' = \frac{rd}{r - R} \tag{5}$$

where, $\alpha$ is the angle, that the refracted marginal ray from the first lens makes with the optical axis.

Substituting the $R$:

$$s' = \frac{d}{1 - (B/r) \sqrt{1 - \exp(-2r^2 / w_0^2)}} \tag{6}$$

$s_0^{\prime}$ of the paraxial focus $F$ of the first lens can be found by considering the limit as $r$ approaches zero with Taylor expansion ($\exp(-ar^2) \approx 1 - ar^2$).

$$s_{0}^{\prime} = \lim_{r \to 0} s^{\prime} \approx \frac{d}{1 - \sqrt{2}B / w_{0}} \tag{7}$$

Then, the longitudinal aberration (LA) $\Delta s^{\prime}$ to be introduced by the first aspheric lens is given by

$$\Delta s^{\prime} = s^{\prime} - s_{0}^{\prime} = \frac{d}{1 - (B / r)\sqrt{1 - \exp(-2r^{2} / w_{0}^{2})}} - \frac{d}{1 - \sqrt{2}B / w_{0}} \tag{8}$$

Expanding Equation 8 to terms including $r^2$, the function of LA can be written as follows:

$$\text{LA} = \Delta s' \approx Ar^2 \tag{9}$$

where:

$$A = \frac{Bd}{\sqrt{2}(w_0 - \sqrt{2}B)^2} \tag{10}$$

This mathematical expression is similar to the third-order aberration theory. Thus, the third-order aberration theory can be used to design laser beam shapers. (**This equation is wrong**)

Equation 9 is thus a parabola approximation of Equation 8:

We can see that in the range of input beam radii from 0 to $\omega_0$, the function of LA can be approximated by parabolic function with very high precision. The parabola and function LA are crossing at the point equal to the **wasit radius $\omega_0$**

**Third-order Aberration Theory**

According to the third-order aberration theory, for an infinitely remote object the transverse spherical aberration $\Delta y_\text{III}^{\prime}$ of an optical system with focal length $f^{\prime}$ is written as:

$$ \Delta y'_{\text{III}} = - \left[ \frac{f'}{2r} \right] S_{\text{I}} \tag{11} $$

where $r$ is the height of a ray, and $S_{\text{I}}$ is the sum of the first Seidel coefficient.

The Longitutional Aberration can be given by:

$$ \frac{\Delta s'_{\text{III}}}{\Delta y'_{\text{III}}} = \frac{f'}{r} \Rightarrow \Delta s'_{\text{III}} = \frac{f'}{r} \Delta y'_{\text{III}} = - \left[ \frac{f'^2}{2r^2} \right] S_{\text{I}} \tag{12} $$

And

$$ S_{\text{I}} = \frac{r^4(n^2+k)}{f'^3(n-1)^2} \tag{13} $$

**For the first aspherical surface**

$$ S_{\text{I1}} = \frac{r^4(n^2 + k_1)}{f'^3_1 (n-1)^2} \tag{14} $$

The first aspherical surface introduces spherical aberration sufficient to create the necessary ray mapping function. Combining Equarion 14 and 12, we have the third order LA of the first lens:

$$\Delta s'_{\text{III1}} = -\frac{n^2+k_1}{2f'_1(n -1)^2} r^2 \tag{15} $$

This aberration of the first aspherical lens can be used to approximate the ray mapping function, assuming:

$$ \Delta s' = \Delta s'_{\text{III1}} \tag{16} $$

Now, we can solve the conic constant of the first aspherical surface $k_1$ under the parabola approximation:

$$ k_1 = \frac{-2\Delta s' f'_1 (n - 1)^2}{r^2} -n^2 \tag{17} $$

Besides, it's convenient to consider the LA $\Delta s'_{\omega_0}$ corresponding to the ray height at the beam waist of radius $\omega_0$. And **for a plano-convex lens**:

$$ f'_1 = s'_0 \tag{18} $$

Thus, the final expression for the conic constant:

$$ k_1 = \frac{-2\Delta s'_{\omega_0} s'_0 (n - 1)^2}{r^2} -n^2 \tag{17} $$

And the vertex radius $r_{c1}$ of the aspheric surface is:

$$ r_{c1} = f'_1(n - 1) = s'_0(n - 1) \tag{18} $$


**For the second aspherical surface**

$$ S_{\text{I2}} = - \frac{R^4(n^2 + k_2)}{f'^3_2 (n - 1)^2} \tag{19} $$

As there should be no wavw aberration present in the plane wave front as it leaves the system. So the spherical aberration of the the whole laser beam shaper must be equal to 0:

$$ S_{\text{I}} = S_{\text{I1}} + S_{\text{I2}} = 0 \tag{20} $$

which gives:

$$ \frac{r^4(n^2 + k_1)}{f'^3_1(n-1)^2} = \frac{R^4(n^2 + k_2)}{f'^3_2(n-1)^2} \tag{21} $$

With:

$$ \frac{r}{R} = \frac{f'_1}{f'_2} \quad f'_1 = s'_0 \quad f'_2 = s'_0 - d $$

The conic constant of the second aspherical surface can be derived:

$$ k_2 = \frac{s'_0k_1 + n^2d}{s'_0 - d} \tag{22}$$

And the vertex radius:

$$ r_{c2} = (n-1)(d - s'_0) \tag{23}  $$

## **Constant OPL Condition**

Imaging a **Galilean Beam Shaper**, whose first lens is a plano-concave lens with thickness $t_{1}$ and sag $z(r)$, and second lens is a convex-plano lens with a thickness $t_{2}$ and sag $Z(R)$, and the distance between the two lenses is $t$.

According to constant OPL condition, a ray at height 0 should have the same OPL with a ray at height $r$ which will arrive the exit pupil at height $R$:

$$ \text{OPL}_0 = \text{OPL}_r \tag{3.1}$$

where

$$ \text{OPL}_0 = nt_1 + t + nt_2 \tag{3.2} $$

and 

$$ \text{OPL}_r = n(t_1 + z(r)) + L + n(t_2 - Z(R)) \tag{3.3}$$

By subsittuting the Equations 3.2 and 3.3 in the Equation 3.1, we will have:

$$ L = t - nz(r) + nZ(R) = nt + nZ(R) - nz(r) + t - nt = n(t + Z(R) - z(r)) + t(1-n) \tag{3.4} $$

The ray at height $R$ after the first lens forms an angle $\theta$ with the optical axis, which gives the following relations:

$$ L\sin\theta = R - r \tag{3.5} $$

$$ L\cos\theta = t - z(r) + Z(R) \tag{3.6} $$

By subsituting the Equation 3.6 in the Equation 3.4, we have:

$$ L = nL\cos\theta + t(1-n) \Rightarrow L(1 - n\cos\theta) = t(1-n) \tag{3.7} $$

Eliminatiing the $L$ with the Equation 3.5:

$$ \frac{1 - n\cos\theta}{\sin\theta} = \frac{t(1-n)}{R-r} \tag{3.8} $$

Square both side and add $(n^2 - 1)$ to both side for simplification, we finally have:

$$ \frac{t^2(1-n)^2}{(R-r)^2} + (n^2 - 1) = \left( \frac{\cos\theta - n}{\sin\theta} \right)^2 \tag{3.9} $$

## **From the Aberration Theory Results**

```python
import numpy as np

class BeamShaper:
    """
    This class is to model a Keplerian or Galilean beam shaper.

    This implementation is based on the formula provided in:
    - Title: "Laser Beam Shaping: Theory and Techniques"
      Editor: Fred M. Dickey
      Chapter: 7
      Edition: 2nd

    Args:
        omega_0 (float): The radius of the input Gaussian beam waist.
        R_max (float): The maximum output radial coordinate.
        r_max (float): The maximum input radial coordinate.
        d (float): The thickness of the first lens' second surface.
        n (float): The refractive index of the two lens.
        type (str): Either "Keplerian" or "Galilean", indicatinig the type of beam shaper.
        approx (bool): indicates if the Longitudinal Aberration is approximated by parabolic
            function and Taylor expansion.
    """

    def __init__(
        self,
        omega_0: float,
        R_max: float,
        r_max: float,
        d: float,
        n: float,
        type: str = "Keplerian",
        approx: bool = True,
    ):
        self.omega_0 = omega_0
        self.R_max = R_max
        self.r_max = r_max
        self.d = d
        self.n = n

        if type not in ["Keplerian", "Galilean"]:
            raise ValueError("type must be either 'Keplerian' or 'Galilean'")
        self.type = type

        self.approx = approx

    @property
    def _B(self) -> float:
        """
        Returns a intermediate coefficient B given by Equation 7.13.
        """
        denom = np.sqrt(1 - np.exp(-2 * (self.r_max / self.omega_0) ** 2))
        if self.type == "Keplerian":
            return -self.R_max / denom
        else:
            return self.R_max / denom

    def s_prime(self, r) -> float:
        """
        Returns the object distance after the first lens at different ray height,
        given by Equation 7.16.
        """
        sqrt = np.sqrt(1 - np.exp(-2 * r**2 / self.omega_0**2))
        s_prime = self.d / (1 - (self._B / r) * sqrt)
        return s_prime

    @property
    def s_prime_zero(self) -> float:
        """
        Returns paraxial object distance, which is also equal to effective focal length f_1,
        given by Equation 7.17 and 7.18.
        """
        if self.approx:
            return self.d / (1 - np.sqrt(2) * self._B / self.omega_0)
        else:
            return self.s_prime(1e-15)

    @property
    def Delta_s_prime(self) -> float:
        """
        Returns the difference between real object distance and paraxial object distance,
        which is actually Longitudinal Aberration (LA), given by Equation 7.19, 7.20 and 7.21.
        """
        return self.s_prime(self.omega_0) - self.s_prime_zero

    @property
    def k_1(self) -> float:
        """
        Returns the conic constant of the first aspherical surface.
        """
        coeff = -2 * self.Delta_s_prime * self.s_prime_zero * (self.n - 1) ** 2
        return coeff / (self.omega_0**2) - self.n**2 

    @property
    def r_c1(self) -> float:
        """
        Returns the vertex radius of the first aspherical surface.
        """
        return -self.s_prime_zero * (self.n - 1)

    @property
    def k_2(self) -> float:
        """
        Returns the conic constant of the second aspherical surface.
        """
        coeff = self.s_prime_zero * self.k_1 + self.n**2 * self.d
        return coeff / (self.s_prime_zero - self.d)

    @property
    def r_c2(self) -> float:
        """
        Returns the vertex radius of the second aspherical surface.
        """
        return (self.n - 1) * (self.d - self.s_prime_zero)

    @property
    def xpan_ratio(self) -> float:
        """
        Returns the expansion ratio of the beam shaper.
        """
        f_1 = self.s_prime_zero
        f_2 = self.d - self.s_prime_zero
        return np.abs(f_2 / f_1)

    @property
    def apodization_factor(self) -> float:
        """
        Returns the apodization factor of the beam shaper.
        """
        return (self.r_max / self.omega_0)**2
    
    @property
    def result(self) -> dict:
        """
        Returns a dictionary containing parameters.
        """
        return {
            "r_c1": float(self.r_c1),
            "r_c2": float(self.r_c2),
            "k_1": float(self.k_1),
            "k_2": float(self.k_2),
        }
```

```python
from pprint import pprint

keplerian = BeamShaper(
    omega_0=2.366,
    R_max=4.05,
    r_max=4.05,
    d=150.0,
    n=1.46071,
    type="Galilean",
    approx=True,
)

pprint(keplerian.result)
```
