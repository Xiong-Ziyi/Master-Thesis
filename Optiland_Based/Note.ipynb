# **Gaussian Beam**

## **Definition**

The Gaussian is a radially symmetrical distribution whose amplitude variation is given by:

$$A(r) = \exp(-\frac{r^2}{\omega_0^2}) \tag{1} $$

where $r$ is defined as the distance from the center of the beam, and $\omega_0$ is the radius at which the amplitude is $1/e$ of its value on the axis.

<img src = "Figures/Gaussian_Beam.jpg" style="display: block; margin: auto; transform: scale();">

The beam size is a function of the distance from the waist. Zemax uses the width or radius to describe beam waist:

$$ \omega(z) = \omega_0 \left[ 1 + \left( \frac{z}{z_R} \right)^2 \right]^{\frac{1}{2}} $$

For large distance the beam size expands linearly, The divergence angle $\theta$ of the beam is given by:

$$ \theta = \frac{\lambda}{\pi n \omega_0} \quad \text{for} \quad z \gg z_R $$

where $z_R$ is the Rayleigh range of the beam given by:

$$ z_R = \frac{\pi \omega_0^2 n}{\lambda} $$

The wavefront's radius of curvature of the beam is a function of the distance from the beam waist:

$$ R(z) = z + \frac{z_R^2}{z} $$

## **Apodization Factor**

When the light source in Zemax is Gaussian laser beam, the apodization matters. 

The Apodization Factor determines what the Gaussian beam's intensity profile looks like across the aperture (or the entrance pupil of the system). More specifically, it determines how fast a Gaussian beams

The light amplitude of a Gaussian beam is:

$$ A(\rho) = \exp(-G\rho^2) \tag{2} $$

Where $A$ and $\rho$ (radial pupil coordinate) are normalized parameters, they are both equal to $0-1$ within their full range. And $G$ is the **apodization factor**.

The intensity of a Gaussian beam is given by:

$$ I(\rho) = A(\rho)^2 = \exp(-2G\rho^2) \tag{3} $$

For example, if we want the Gaussian beam is clipped by the system clear aperture right at $13.5\%$ ($1/e^2$) of the peak value:

$$ \frac{I(1)}{I(0)} = \frac{1}{e^2} = \frac{e^{-2G \times 1^2}}{e^{-2G\times 0}} \Rightarrow G = 1 \tag{4} $$

If we write the Gaussian profile with normalized radial coordinate ($\rho = r/r_{max}$):

$$ A(\rho) = \exp(-\frac{r_{max}^2}{\omega_0^2}\rho^2) \tag{5} $$

which also gives Apodization factor:

$$ G = \frac{r_{max}^2}{\omega_0^2} \tag{6}$$

# **Intensity and Irradiance**

**Irradiance (E)** is the optical power arriving at a surface per unit area:

$$ E = \frac{d\Phi}{dA} $$

$\Phi$: radiant flux ($W$); $A$: area ($m^2$).

**Intensity (I)** is the power emitted per unit solid angle.

$$ I = \frac{d\Phi}{d\Omega} $$

# **Cartesian Oval Collimator**

Consider a point source at $z = 0$, and at $z = z_0$ is there a surface whose sag is $z(r)$. The divergent rays from the source will be collimated by the surface. Then the sag of this surface is a **Cartesian Oval**.

<img src = "Figures/Cartesian_Oval.png" style="display: block; margin: auto; transform: scale();">

The sag of this Cartesian Oval can be determined by Optical Path Length (OPL) condition:

$$ \sqrt{r^2 + (z(r) + z_0)^2} = z_0 + nz(r) $$

Square both side and rearrange:

$$ r^2 = (n^2 - 1)z^2(r) + 2 z(r) z_0 (n-1)  $$

Divide both side with $n^2 - 1$ and add an additional term:

$$ \frac{r^2}{n^2 -1} + \frac{z_0^2}{(n + 1)^2} = z^2(r) + \frac{2z_0}{n + 1}z(r) + \frac{z_0^2}{(n + 1)^2} $$

Then we have:

$$ \frac{r^2}{n^2 -1} + \frac{z_0^2}{(n + 1)^2} = \left[z(r) + \frac{z_0}{n + 1}\right]^2 $$

Finally:

$$ z(r) = \sqrt{\frac{r^2}{n^2 -1} + \frac{z_0^2}{(n + 1)^2}} - \frac{z_0}{n + 1} $$

Let:

$$ B = \frac{1}{n^2 -1} \quad A = \frac{z_0}{(n + 1)} $$

Then:

$$ z(r) = \sqrt{Br^2 + A^2} - A $$

For small $r$, $z(r)$ can be approximated by Taylor expansion:

$$ z(r) = A \sqrt{1 + \frac{Br^2}{A^2}} - A \approx A(1 + \frac{Br^2}{2A^2}) - A = \frac{Br^2}{2A}$$

Then:

$$ z(r) \approx \frac{r^2}{2z_0(n-1)} $$

For a standard surface:

$$ z(r)' = \frac{cr^2}{1 + \sqrt{1 - (k+1)c^2r^2}} = \frac{cr^2}{(k+1)c^2r^2} (1 - \sqrt{1 - (k+1)c^2r^2}) $$

Similarly, for small $r$:

$$ z(r)' \approx \frac{cr^2}{(k+1)c^2r^2} \left[1 - \left(1 - \frac{(k+1)}{2}c^2r^2\right)\right] $$

So: 

$$ z(r)' \approx \frac{cr^2}{2} = \frac{r^2}{2R} $$

Compare with $z(r)$, we can get the vertex radius:

$$ R = z_0(n-1) $$

Conic constant $k$ is is given by:

$$ k = -e^2 $$

where $e$ is **eccentricity**

And $e$ is defined by the distance between a point ($P$) at the curve to the focus of the curve $|PF|$ over the distance between $P$ to directrix $\|PP'|$:

$$ e = \frac{|PF|}{|PP'|} $$

<img src = "Figures/Eccentricity.png" style="display: block; margin: auto; transform: scale();">

In this case, $\text{|PF|}$ is the distance from the point source to a point on the Cartesian Oval:

$$ |PF| = \sqrt{r^2 + (z(r) + z_0)^2} $$

and the directrix is a line perpendicular to the optical axis at $z_{ref}$, so the distance from the point $\text{P}$ to the directrix is:

$$ |PP'| = z_{ref} - (z_0 + z(r)) $$

For a general Cartesian Oval, a random ray goes from focus $F_1$ to the surface point $P$ in medium $n_1$, and goes from $P$ to focus $F_2$ in medium $n_2$. The definition of Cartesian Oval says that the OPL of any any random rays must be a constant:

$$ \text{OPL} = n_1 |PF_1| + n_2|PF_2| = \text{const} $$

When $|PF_2|=\infty$ in this case, a reference plane is chosen, which also plays the role as directrix in defining eccentricity, results in:

$$ \text{OPL} = n_1|PF| + n_2|PP'| = \text{const} $$

Rearrange: 

$$ -\frac{n_1}{n_2}||PF| = \left(z_{ref} - \frac{\text{const}}{n_2}\right) - (z_0 + z(r))$$

Since $z_{ref}$ can be anywhere and $\text{const}/n_2$ is also a constant, so we can set a new directrix with a distance to the point $P$:

$$ |PD| =  -\left(z_{ref} - \frac{\text{const}}{n_2}\right) + (z_0 + z(r))$$

which gives us:

$$ -\frac{n_1}{n_2}||PF| = -|PD| \quad \Rightarrow \quad e = \frac{|PF|}{|PD|} = \frac{n_2}{n_1} $$

And the conic constant $k$ becomes:

$$ k = -e^2 = -\left(\frac{n_2}{n_1}\right)^2 $$
