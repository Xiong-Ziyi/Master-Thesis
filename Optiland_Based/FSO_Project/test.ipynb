# **Beam Shaper**

```python
from pprint import pprint
from classes.beam_shaper import BeamShaper
  
beamshaper_np = BeamShaper(
    omega=2.366,
    R_max=4.05,
    r_max=4.05,
    d=150.0,
    n=1.46071,
    type="Galilean",
    approx=True,)

pprint(beamshaper_np.result)
```

```python
import torch

beamshaper_torch = BeamShaper(
    omega=2.366,
    R_max=4.05,
    r_max=4.05,
    d=150.0,
    n=1.46071,
    type="Galilean",
    approx=True,
    backend="torch",
    device="cuda" if torch.cuda.is_available() else "cpu"
)

pprint(beamshaper_torch.result)
```

# **Super Gaussian**

```python
from classes.super_gaussian import SuperGaussianIrradiance

supergaussian_np = SuperGaussianIrradiance(
    peak_irradiance=1,
    flattop_radius=8,
    steepness=10,
    max_coord=10, 
    num_samples=100)

supergaussian_np.plot_cross_section()
supergaussian_np.plot_2d()
```

```python
import torch

supergaussian_torch = SuperGaussianIrradiance(
    peak_irradiance=1,
    flattop_radius=8,
    steepness=10,
    max_coord=10, 
    num_samples=100,
    backend='torch',
    device='cuda' if torch.cuda.is_available() else 'cpu'
)

supergaussian_torch.plot_cross_section()
supergaussian_torch.plot_2d()
```
